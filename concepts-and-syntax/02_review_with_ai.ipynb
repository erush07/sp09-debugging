## AI Setup Prompt

Start by pasting this prompt into your AI tool to guide the interaction:

```
I am a beginner in an introductory Python programming course. When I ask questions or describe a concept, please respond as my tutor by explaining in simple, easy-to-understand language. Correct any mistakes gently and clarify misconceptions. Use examples or metaphors when helpful. Let me know what I'm understanding well and what I might need to improve or clarify.
```



## Questions to Review Concepts

Ask your AI tutor these questions to improve your understanding of debugging:

- What is a **syntax error** and how do I recognize one in Python?
- What is a **runtime error**, and how is it different from a syntax error?
- What is a **logic error**, and why can it be tricky to spot?
- When my python code crashes, the terminal has a bunch of words and its hard to make sense of it. Can you help me understand how to read what it says when an error occurs?
- Can you give me examples of each type of bug?
- How can I fix a bug if Python isn’t giving me an error message?



## Check Your Understanding

Try explaining these ideas to AI in your own words. Ask it to correct you if you’re wrong:

- What happens when Python encounters a syntax error?
- How is a runtime error different from a logic error?
- What kind of bug is hardest to catch, and why?



## Generate Your Own Practice Question

Use this prompt to have AI create a custom debugging practice problem for you:
